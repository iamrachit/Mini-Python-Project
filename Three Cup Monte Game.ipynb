```python
from random import shuffle
```

```python
mylist = ['', '0', '']
```

```python
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist
```

```python
def player_guess():
    guess = ''
    while guess not in ['0','1','2']:
        guess = input("pick a number : 0 , 1 , 2 ")
    return int(guess)
```

```python
player_guess()
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == '0':
        print("correct")
    else:
        print("wrong guess")
        print(mylist)
```

```python
mylist = ['','0','']

mixedup_list = shuffle_list(mylist)

guess = player_guess()

check_guess(mixedup_list,guess)
```
