```python
def spy_game(nums):
    code = [0,0,7,'x']
    for num in nums:
        if num == code[0]:
            code.pop(0)
            
    return len(code) == 1
```

```python
spy_game([1,2,34,0,0,7,5])
```

```python
spy_game([1,0,2,4,0,5,7])
```

```python
spy_game([1,7,2,0,4,5,0])
```

```python
def count_primes(num):
    # Check for 0 and 1 input
    if num < 2:
        return 0
    # 2 or greater 
    
    primes = [2]
    x = 3
    
    # x will go through all the number 
    
    while x <= num:
        for y in range(3,x,2):
            if x%y == 0:
                x += 2
                break
        else:
            primes.append(x)
            x += 2
            
    print(primes)
    return len(primes)
```

```python
count_primes(100)
```
